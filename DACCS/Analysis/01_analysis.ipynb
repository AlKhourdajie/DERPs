```python
# Load packages and other dependencies
import pyam
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1) Load data

```python
# Load data 
path = "../data/" # change to your path
data_file = pd.read_csv(path +"DACCS_GCAM.csv", encoding='ISO-8859-1')

# pyam data_file
DACCS_df = pyam.IamDataFrame(data_file)
```

```python
timeseries_variables = [
    'Emissions|CO2',
    'Carbon Sequestration|Direct Air Capture',
     'Primary Energy',
    'Final Energy',
     'Temperature|Global Mean',
]
```

```python
# Define a safe function to sanitise the variable name for saving files

def sanitize_filename(filename):
    return filename.replace("|", "_").replace("/", "_").replace("*", "")
```

```python
# For every scenario, plot the emissions of CO2

for variable in timeseries_variables:
    DACCS_df.filter(variable=variable, region = 'World').plot()
    # save plots in the figures folder, named by the variable and region
    sanitized_variable = sanitize_filename(variable)
    plt.savefig(f'../figures/{sanitized_variable}_World.png')
    
```

```python
stakedline_variables = [
    'Primary Energy*',
    'Final Energy*',
    'Emissions*',
    'Land Cover*',
    'Secondary Energy*',
]
```

```python
# List of scenarios you want to plot
scenarios = DACCS_df.scenario

# For every scenario, plot the variables in a stacked line plot with side-by-side panels
for variable in stakedline_variables:
    # Create subplots, one per scenario
    fig, axes = plt.subplots(1, len(scenarios), figsize=(15, 5), sharey=True)
    fig.suptitle(f'{variable} for different scenarios', fontsize=16)
    
    # Variable to store the handles and labels for the legend
    handles, labels = None, None
    
    # Loop through each scenario and create a stacked line plot for that scenario
    for i, scen in enumerate(scenarios):
        ax = axes[i]
        plot = DACCS_df.filter(variable=variable, level=1, scenario=scen, region='World').plot.stack(ax=ax)
        
        # Get the handles and labels from the first plot for the legend
        if handles is None and labels is None:
            handles, labels = ax.get_legend_handles_labels()
        
        # Remove the legend from all plots
        ax.legend().remove() 
        
        ax.set_title(f'Scenario: {scen}')
        ax.set_xlabel('Year')
        ax.set_ylabel(variable)
    
    # Create a single legend below the plots with 6 columns
    fig.legend(handles, labels, loc='lower center', ncol=6, bbox_to_anchor=(0.5, -0.2))
    
    # Adjust the spacing between subplots (reduce the space further)
    plt.subplots_adjust(wspace=0.02)  # Reduce the horizontal space between the subplots even more
    
    # Sanitize the variable name for the filename
    sanitized_variable = sanitize_filename(variable)
    
    # Save the plot with all the scenarios side by side
    plt.tight_layout(rect=[0, 0, 1, 0.93])  # Adjust layout to fit the suptitle and make space for the legend
    plt.savefig(f'../figures/{sanitized_variable}_scenarios.png', bbox_inches='tight')
    plt.show()
```
